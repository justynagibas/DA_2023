```python
import numpy
import matplotlib.pyplot as plt
import numpy as np
import seaborn
from scipy.stats import uniform, norm, beta
```

# Ex 2 Generate samples

```python
uniform_distribution = numpy.random.uniform(size=1000)
gauss_distribution = numpy.random.normal(loc=0, scale=1, size=1000)
gauss_distribution_2 = numpy.random.normal(loc=0, scale=10, size=1000)
beta_distribution_1_1 = numpy.random.beta(a=1, b=1, size=1000)
beta_distribution_2_2 = numpy.random.beta(a=2, b=2, size=1000)
beta_distribution_3_9 = numpy.random.beta(a=3, b=9, size=1000)
norm_dist_2_dim_1 = numpy.random.multivariate_normal(mean=[0,0], cov=[[1,0],[0,1]], size=10000)
norm_dist_2_dim_2 = numpy.random.multivariate_normal(mean=[0,0], cov=[[2,1],[1,2]], size=10000)
```

# Ex 3 Plotted 1D histograms

```python
_, bins, _ =  plt.hist(uniform_distribution, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, uniform.pdf(bins), color='r', label='Analytical distribution')
plt.title(r'Uniform distribution on interval [0;1]')
plt.legend()
plt.show()
```

```python
_, bins, _ = plt.hist(gauss_distribution, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, norm.pdf(bins,loc=0,scale=1), color='r', label='Analytical distribution')
plt.title(r'Normal distribution $\mu$ = 0, $\sigma$ = 1')
plt.legend()
plt.show()
```

```python
_, bins, _ =plt.hist(gauss_distribution_2, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, norm.pdf(bins, loc=0, scale=10), color='r', label='Analytical distribution')
plt.title(r'Normal distribution $\mu$ = 0, $\sigma$ = 10')
plt.legend()
plt.show()
```

```python
_, bins, _ = plt.hist(beta_distribution_1_1, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, beta.pdf(bins, a=1, b=1), color='r', label='Analytical distribution')
plt.title(r'Beta distribution $\alpha$ = 1, $\beta$ = 1')
plt.legend()
plt.show()
```

```python
_, bins, _ = plt.hist(beta_distribution_2_2, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, beta.pdf(bins, a=2, b=2), color='r', label='Analytical distribution')
plt.title(r'Beta distribution $\alpha$ = 2, $\beta$ = 2')
plt.legend()
plt.show()
```

```python
_, bins, _ = plt.hist(beta_distribution_3_9, bins='auto', density=True, label='Numerical distribution')
plt.plot(bins, beta.pdf(bins, a=3, b=9), color='r', label='Analytical distribution')
plt.title(r'Beta distribution $\alpha$ = 3, $\beta$ = 9')
plt.legend()
plt.show()
```

# Ex 4 Plotted 2D distributions

```python
plt.hist2d(norm_dist_2_dim_1[:,0], norm_dist_2_dim_1[:,1], bins=50,  cmap='Blues')
plt.title(r'2D normal distribution $\mu=[0,0]^T$ $\Sigma$=[[1,0],[0,1]]')
plt.show()
```

```python
seaborn.displot(x=norm_dist_2_dim_1[:,0], y=norm_dist_2_dim_1[:,1])
plt.title(r'2D normal distribution $\mu=[0,0]^T$ $\Sigma$=[[1,0],[0,1]]')
plt.show()
```

```python
plt.hist2d(norm_dist_2_dim_2[:,0], norm_dist_2_dim_2[:,1], bins=50, cmap='Blues')
plt.title('2D normal distribution $\mu=[0,0]^T$ $\Sigma$=[[2,1],[1,2]]')
plt.show()
```

```python
seaborn.displot(x=norm_dist_2_dim_2[:,0], y=norm_dist_2_dim_2[:,1])
plt.title('2D normal distribution $\mu=[0,0]^T$ $\Sigma$=[[2,1],[1,2]]')
plt.show()
```

Using not enough bins make plot illegible and don't represent distribution correctly. The same situation is caused by too much bins. Also presenting 2D distribution using seaborn gives better results.

# Ex 5 Points generation

```python
x_plot = []
y_plot = []
alpha_plot = []
beta_plot = []
for i in range(1000):
    alpha = numpy.random.normal(loc=2, scale=0.1)
    beta = numpy.random.normal(loc=0.2, scale=0.2)
    x = numpy.random.normal(loc=0, scale=3)
    mi = alpha + beta * x
    sigma = abs(numpy.random.normal(0, 1))
    y = numpy.random.normal(loc=mi, scale=sigma)
    x_plot.append(x)
    y_plot.append(y)
    beta_plot.append(beta)
    alpha_plot.append(alpha)
```

# Ex 6 Plotting points distribution

```python
x = np.arange(start=-6, stop=6, step=0.1)
for i in range(1000):
    y = alpha_plot[i]+x*beta_plot[i]
    plt.plot(x,y,'r',alpha=0.5)
plt.scatter(x_plot, y_plot, zorder=2, alpha=0.9)
plt.title('Comparison of numerical and analytical distribution')
plt.show()
```
